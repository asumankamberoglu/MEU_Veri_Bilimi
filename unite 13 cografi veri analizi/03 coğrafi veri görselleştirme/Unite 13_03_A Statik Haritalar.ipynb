## Poligon Tipi Veriler

İsveç'in Malmö kentinin mahallelerinin nüfus dağılım verisini (malmo-pop.shp) elimizde bulunduruyoruz. En sonda bulunan geometry sütunu ile ifade edilen mahalle sınırlarının içinde hangi yaş aralığından kaç kişinin yaşadığını bu veri setinde görebilmekteyiz. 

```python
malmo_mahalle = gpd.read_file("/content/drive/MyDrive/Colab Notebooks/Geopandas Data Visualisation/malmo-pop.shp")
malmo_mahalle.head()
```

```python
# Veri setinin boyutlarını irdeleyelim. 
malmo_mahalle.shape
```

```python
# Veri setinin koordinat referans sistemini inceleyelim. 
# Veri setinin kartezyen koordinat sisteminde ve İsveç yerel 1999 datumunda bulunduğunu görebilmekteyiz.
malmo_mahalle.crs
```

```python
# Veri setini kaba olarak çizdirelim. 
malmo_mahalle.plot();
```

```python
# Çizimin boyunu büyütelim. Mahalle sınırları daha görünür olacaktır. 
malmo_mahalle.plot(figsize=(12,10));
```

```python
# Çizimin rengini gri yapalım. 
malmo_mahalle.plot(figsize=(12,10),
                color="gray");
```

```python
# Çizimin şeffaflığını artıralım. Bunun için 'alpha' parametresini kullanalım. 
malmo_mahalle.plot(figsize=(12,10),
                color="gray",
                alpha=0.7);
```

```python
# Mahalle sınırlarını renklendirmek için 'edgecolor' parametresini kullanalım.
# Buraya yazacağımız renk RGB uzayındaki tüm renkleri 6 haneli rakam ve harflerle temsil eden HEX kodlaması ile yazılmalıdır.
# HEX kodunu elde etmek için https://www.google.com/search?q=color+picker bağlantısından renk seçimi yapılması yeterli olacaktır. 
malmo_mahalle.plot(figsize=(12,10),
                color="gray",
                alpha=0.7,
                edgecolor="#CFAC84");
```

```python
# Mahalle sınır çizgilerinin kalınlığını artıralım. Bunun için 'linewidth' parametresi kullanılacaktır. 
malmo_mahalle.plot(figsize=(12,10),
                color="gray",
                alpha=0.7,
                edgecolor="#CFAC84",
                linewidth=2);
```

## Çizgi Tipi Veriler

```python
# Malmö şehrine ait yolları içeren "malmo-street.shp" veri setini kullanalım. 
# Bu veri setinin de son sütunu çizgi tipi geometri içermektedir. 
malmo_yollar = gpd.read_file("/content/drive/MyDrive/Colab Notebooks/Geopandas Data Visualisation/malmo-streets.shp")
malmo_yollar.head()
```
