# <h1><center>Altlık Haritalar (Base Maps) </center></h1>

Kendi veri setlerimizle birlikte görselleştirme amaçlı internetten altlık haritalar da kullanmamız gerekmektedir. Geopandas kütüphanesi ile görselleştirdiğimiz veri setimizin arka planına raster formatında altlık harita yükleyebilmek için Python'un [contextily](https://github.com/geopandas/contextily) kütüphanesini kullanabiliriz. Bu kütüphanenin sunduğu altlık haritalar **WGS84** (EPSG:4326) ya da **Küresel Merkator** (EPSG:3857) projeksiyonunda olduğu içinde, elimizdeki veri setlerinin Geopandas ile bu projeksiyonlara dönüştürülmesi gerekmektedir. 

```python
#!pip install geopandas
#!pip install contextily
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx
```

```python
# Anaokulları veri setimizi yükleyelim.
malmo_anaokullari= gpd.read_file("/content/drive/MyDrive/Colab Notebooks/Geopandas Data Visualisation/preschools.shp")
```

```python
# Bu veri setini Geopandas ile küresel merkator projeksiyonuna dönüştürerek görselleştirelim. Gerekli parametreleri kullanalım.

fig, ax = plt.subplots(figsize=(12, 10))
malmo_anaokullari.to_crs(epsg=3857).plot(ax = ax,
                figsize=(12,12),
                markersize=40,
               color="black",
               edgecolor="white",
               alpha=0.8,
               marker="o"
            );


```

```python
# Contextily kütüphanesinin sunduğu altlık harita çeşitlerinin isimlerini ekrana verelim. 
ctx.providers.keys()
```

contextily_providers.png


```python
# Malmö anaokulları veri setini Stamen.TonerLite altlık haritası ile görselleştirelim.
fig, ax = plt.subplots(figsize=(12, 10))
malmo_anaokullari.to_crs(epsg=3857).plot(ax = ax,
                figsize=(12,12),
                markersize=40,
               color="black",
               edgecolor="white",
               alpha=0.8,
               marker="o"
            );
ctx.add_basemap(ax, url=ctx.providers.Stamen.TonerLite);
```

```python
# Malmö anaokulları veri setini OpenStreetMap.Mapnik altlık haritası ile görselleştirelim.
fig, ax = plt.subplots(figsize=(12, 10))
malmo_anaokullari.to_crs(epsg=3857).plot(ax = ax,
                figsize=(12,12),
                markersize=40,
               color="black",
               edgecolor="white",
               alpha=0.8,
               marker="o"
            );
ctx.add_basemap(ax, url=ctx.providers.OpenStreetMap.Mapnik);
```

```python
# Malmö anaokulları veri setini CartoDB.DarkMatter altlık haritası ile görselleştirelim.
fig, ax = plt.subplots(figsize=(12, 10))
malmo_anaokullari.to_crs(epsg=3857).plot(ax = ax,
                figsize=(12,12),
                markersize=40,
               color="black",
               edgecolor="white",
               alpha=0.8,
               marker="o"
            );
ctx.add_basemap(ax, url=ctx.providers.CartoDB.DarkMatter);
```
