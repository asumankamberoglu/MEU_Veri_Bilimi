```python
malmo_mahalle = gpd.read_file("/content/drive/MyDrive/Colab Notebooks/Geopandas Data Visualisation/malmo-pop.shp")
malmo_yollar = gpd.read_file("/content/drive/MyDrive/Colab Notebooks/Geopandas Data Visualisation/malmo-streets.shp")
malmo_anaokullari = gpd.read_file("/content/drive/MyDrive/Colab Notebooks/Geopandas Data Visualisation/preschools.shp")
```

### Yan Yana Grafikler (Matplotlib Hatırlatma) 

```python
fig, ax  = plt.subplots(3, figsize=(14,12))
```

```python
fig, ax  = plt.subplots(1, 3, figsize=(14,12))
```

```python
# Her bir veri setini ayrı ayrı aynı çerçevede çizdirmek istediğimizde ax=ax[n] sözdizmini kullanırız. "n" buradaki alt grafik indeksidir. 
fig, ax = plt.subplots(1, 3, figsize=(20,18))
malmo_mahalle.plot(ax=ax[0], figsize=(18,14))
malmo_yollar.plot(ax=ax[1], figsize=(18,14))
malmo_anaokullari.plot(ax=ax[2], figsize=(18,14))
```

### Üst Üste Bindirilen Çizimler (Çok Katmanlı - Overlay)
Veri setlerimiz aynı koordinat referans sisteminde ise üst üste bindirilerek çizilebilir. Bu seçenek, farklı veri setlerinden oluşan çok katmanlı tematik haritaların oluşturulması için önemlidir. 
Yan yana çizimden farklı olarak, bu çizimde aynı çerçeveyi kullanacağız. Dolayısıyla, `ax = ax` sözdizimi her çizim için sabit kalacak, herhangi bir indeks girişi yapmayacağız. 

```python
fig, ax = plt.subplots(1,1,  figsize=(12,10))
malmo_yollar.plot(ax=ax, color='#C05A3C', linewidth=0.3, alpha=0.5);
malmo_mahalle.plot(ax= ax,
                color='#F1ECEB',
                alpha=0.7);
malmo_anaokullari.plot(ax = ax,
                figsize=(12,12),
                markersize=40,
               color="black",
               edgecolor="white",
               alpha=0.8,
               marker="o");


plt.title('İsveç Malmo Mahalleleri', fontsize=30, fontname="DejaVu Sans", color="grey")
ax.axis("off")
plt.axis('equal')
plt.show()
```
