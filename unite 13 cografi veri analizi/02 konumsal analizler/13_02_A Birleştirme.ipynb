# <h1><center>Birleştirme İşlemleri</center></h1>
Bu alt bölümde, veri setlerini birleştirmeyi irdeleyeceğiz. Veri setleri hem tablo olarak, hem de konumsal olarak birleştirilebilir. İkisinin nasıl ve ne zaman kullanılacağını görelim. Öncelikle kütüphanelerimizi yükleyelim.

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# ülkeler veri setini okuyalım.
ulkeler = gpd.read_file('countries.shp')
ulkeler.head()
```

```python
# Çatışmalar veri setinin ilk 500 satırını okuyalım.
catismalar = gpd.read_file('acled2019.shp', rows = 500)
catismalar.head()
```

## 1. Tablo Birleştirme

Ülkeler veri setini, seçtiğimiz sütunlardan oluşacak şekilde iki ayrı tabloya bölelim.

```python
ulkeler_geometri = ulkeler[['geometry', 'NAME']]
ulkeler_geometri.head()
```

Hatırlanacağı üzere, GeoDataFrame veri tipi geometry sütununu barındırır.

İkinci tablomuz ise sadece ülke ismi (NAME) ve Tahmin edilen nüfus (POP_EST) sütunlarını içersin.

```python
ulkeler_isim=ulkeler[['NAME', 'POP_EST']]
ulkeler_isim.head()
```

Bu tablo ise, geometry sütunu içermediği için GeoDataFrame olmayacaktır.

**Oluşturduğumuz bu iki tabloyu, ortak sütun olan isim (NAME) sütununu kullanarak birleştirebiliriz. Bu işlem pandas kütüphanesi ile yapılan bir işlemdir ve herhangi bir konumsal operasyon içermez.**

```python
ulkeler_gdf_birlesik=ulkeler_geometri.merge(ulkeler_isim, on ='NAME', how ='left')
ulkeler_gdf_birlesik.head()
```

```python
type(ulkeler_gdf_birlesik)
```

```python
ulkeler_df_birlesik = ulkeler_isim.merge(ulkeler_geometri, on ='NAME', how ='left')
ulkeler_df_birlesik.head()
```

```python
type(ulkeler_gdf_birlesik)
```

## 2. Konumsal Birleştirme (Spatial Join)
Yukarıdaki tabloların birleştirilmesi işleminde herhangi bir konumsal ilişki bulunmamaktadır. Konumsal olarak birleştirme işleminin gerçekleştirilebilmesi için iki tabloda da geometry sütunun olması ve verilerin konumlarına göre birleşmesi beklenir.

```python
ulkeler_geometri.head()
```

Çatışmalar veri setindeki olay kodu (event_id_c), olayın tarihi (event_date), kayıp sayısı (fatalities) ve geometry sütunlarını ayırarak yeni bir GeoDataFrame oluşturalım.

```python
catismalar_gdf = catismalar[['event_id_c', 'event_date','fatalities', 'geometry']]
catismalar_gdf.head()
```

```python
type(catismalar_gdf)
```

Ülkeler ve çatışmalar veri setlerini konumsal olarak birleştireceğiz. Bu iki tablonun içerdiği konumsal veriye göre birleştirme işlemi gerçekleştirilecektir. Birleştirme işleminden sonra, çatışmaların gerçekleştiği yerlerin, hangi ülke sınırları içinde kaldığı da anlaşılabilecektir. Geopandas kütüphanesinde konumsal birleştirme (spatial join) işlemi için `sjoin()` fonksiyonu kullanılır. 

```python
ulkelerdeki_catismalar = gpd.sjoin(catismalar_gdf, ulkeler_geometri, how="inner", op='within')
ulkelerdeki_catismalar.sample(10)
```

Çatışmanın yaşandığı her noktanın, ülke verisi ile konumsal olarak birleştirildiğini görmektesiniz. `op` parametresinin `within` olarak seçilmesi, her çatışma noktasının ait olduğu ülke poligonu ile birleştirilmesini sağlamaktadır.
