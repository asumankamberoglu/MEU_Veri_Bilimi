## 2.1 Çizgi Grafikler

Çok basit bir çizgi grafikle, matplotlib örneklerine başlayalım. 

```python
yil =  [1950, 1970, 1990, 2010] # Yıllardan oluşan listemizi oluşturalım. 
nufus =  [2.519, 3.692, 5.263, 6.972] # O yıllardaki dünya nüfusunu barındıran listeyi de oluşturalım.
```

`plt.plot(x,y)` komutu ile grafiği çizdirme işlemini başlatıyoruz. 
Parantez içinde x ile belirtilen yere, x ekseninde görmek istediğimiz veri setini; y ile belirtilen yere ise, y eksenin görmek istediğimiz veri setini yazıyoruz. Çizdireceğimiz bu grafikte, x ekseninde `yil` listesini, y ekseninde `nufus` listesini görmeyi arzu ediyoruz. 
`plt.show()` komutu ile de grafiğimizi ekrana veriyoruz.Dolayısıyla; 

```python
plt.plot(yil, nufus) 
plt.show()
```

Görüldüğü üzere grafiğimiz çizdirildi. Ancak, oluşan bu grafik çok fazla sayıda eksiğe sahip. Dolayısıyla grafiğimizin anlatımını zenginleştirmemiz gerekmektedir. 

### 2.1.1 Eksen İsimleri (Label)
X eksenine isim verebilmek için `plt.xlabel('string')` komutunu; Y eksenine isim verebilmek için de `plt.ylabel('string')` komutunu kullanıyoruz. String diye adlandırılan parantez içine, isim olarak atamak istediğimiz stringi yazıyoruz. 

```python
plt.plot(yil, nufus) 
plt.xlabel('Yıllar')
plt.ylabel('Dünya Nüfusu')
plt.show()
```

### 2.1.2 Çizgi Rengi (color), Çizgi Kalınlığı (width) ve Çizgi Stili (style)
Çizgi grafiğinde, çizginin rengini default olarak gelen gök mavisi renginden farklı olarak kullanmak isteyebiliriz. Aynı şekilde çizgi kalınlığını artırabilir ya da azaltabilir. Çizginin dümdüz bir çizgi olması yerine, kesik kesik ya da nokta nokta olmasını isteyebiliriz. 

Bu parametreleri `plt.plot(x,y, color = "renk", linewidth = "çizgi kalınlığı", linestyle = "çizgi stili")` komutu içinde argüman olarak istediğimiz sayıda kullanabiliriz. Bunlardan birini ya da hepsini ayarlama olanağımız bulunmaktadır.

Örneğin;


```python
plt.plot(yil,nufus, color = "red", linewidth = "2.5", linestyle = "--")
plt.xlabel('Yıllar')
plt.ylabel('Dünya Nüfusu')
plt.show()
```

Görüldüğü üzere, çizgi grafiğimizin çizgisi kırmızı renge döndü, çizgisi daha da kalınlaştır ve çizginin kesik kesik olması sağlandı. Aşağıdaki tabloda, matplotlib'te kullanabileceğiniz sabit renklerin isimlerini görebilirsiniz. 
![figure](https://matplotlib.org/2.0.0/_images/named_colors.png)

Not: En temel renkler için sadece baş harfleri de kullanabilirsiniz. Ancak grafiklerinizin en estetik halini üretebilmek için ara renkleri de sık sık kullanmanızda fayda var. 
Örneğin: Siyah için `'k'`, beyaz için `'w'`, mavi için `'b'`, kırmızı için `'r'`, yeşil için `'g'`. 


Aşağıda da temel çizgi stillerini nasıl kodlayacağımızı gösteren bir şekil sunulmuştur. Bu dördünü akılda tutmakta fayda var.
![figure](https://i.hizliresim.com/kxBXMP.png)

Benzer şekilde, çok daha detaylı çizgi stillerini gösteren şekil aşağıda sunulmuştur. Bu şeklin ikinci kısmında gösterilen çizgi stillerinin karşılık geldiği sayısal değeri komut içerisindeki argümana girmeniz gerekmektedir. 

Bunları ezberlemenize gerek yoktur. Fazlasıyla grafik gösterime ihtiyaç duyduğunu çalışmalarda bu renk ve çizgi stili tablolarını baş ucunda bulundurmanızda fayda vardır. 

![figure](https://matplotlib.org/3.0.2/_images/sphx_glr_linestyles_001.png)

Başka bir örnekle devam edelim.
**[-π,+π]** kapalı aralığı içinde 256 adet eşit aralıkla sıralanmış değeri bir NumPy dizisi olarak atayalım ve bu değerler üzerinde grafik çalışmaları yapalım.

```python
import numpy as np

X = np.linspace(-np.pi, np.pi, 256, endpoint=True) 
```

`np.linspace` fonksiyonu için belirlediğimiz aralığın uçlarının kapalı olduğunu belirtmek için `endpoint = True` argümanını kullandık.

Şimdide bu değerlerin sinüs ve kosinüs değerlerini hesaplatalım. 

```python
C, S = np.cos(X), np.sin(X) 
# Bu şekilde yazım, C adında kosinüs değerlerinden oluşan listeyi, S adında da sinüs değerlerinden oluşan listeyi oluşturur.
```

Bu grafikte iki adet değeri çizdireceğiz. X olarak belirlediğimiz [-π,+π] kapalı aralığı X ekseninde sabit olmak üzere; Y ekseninde hem C hem de S dizilerini çizdireceğiz. Bunun için;

```python
plt.plot(X,C)
plt.plot(X,S)
plt.show()
```

```python
#Kosinüs fonskiyonunu mavi, sinüs fonksiyonunu yeşil ile çizdirelim ve her ikisinde de kesiksiz çizgi ve 1 piksel kalınlık kullanalım.
plt.plot(X, C, color="blue", linewidth=1.0, linestyle="--")

plt.plot(X, S, color="green", linewidth=1.0, linestyle="--")
plt.show()
```

### 2.1.3 Eksen Aralığı (limits)
Aynı örnekten devam ederek, x eksenini [-4,4] aralığında, y eksenini de [-1,1] aralığında çizdirelim. Bunun için `plt.xlim(Aralık)` ve `plt.ylim(Aralık)` komutları kullanılır.


```python
plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")
plt.plot(X, S, color="green", linewidth=1.0, linestyle="-")
plt.xlim(-4.0,4.0)
plt.ylim(-1.0,1.0)
plt.show()
```

Ancak elimizde bulunan veriyi kendi belirlediğimiz bir eksen aralığı içerisinde çizdirmek her zaman güzel grafikler oluşturmaz. Elimizdeki verinin bulunduğu değer aralığına uygun bir eksen aralığı belirlemekte her zaman fayda vardır. Aksi takdirde, grafik çok sıkışabilir ya da belirli bir yerde toplanabilir. Örneğin;

```python
plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")
plt.plot(X, S, color="green", linewidth=1.0, linestyle="-")
plt.xlim(X.min()*1.1, X.max()*1.1) # Elimizdeki verinin min ve max değerlerinin 1.1 katı olan değerlerde X eksenini aralıklandırıyoruz. 
plt.ylim(C.min()*1.1, C.max()*1.1) # Elimizdeki verinin min ve max değerlerinin 1.1 katı olan değerlerde Y eksenini aralıklandırıyoruz. 
plt.show()
```

### 2.1.4 Eksen İşaretleri (ticks)
Eksenler üzerindeki işaretlemek istediğimiz değerleri, aynı örneği kullanarak belirleyelim. 

Kullandığımız aralık radyan cinsinde aralıklara sahip olduğu için, bu trigonometrik fonksiyonun x ekseni işaret değerlerini pi sayısının katları olarak belirlemekte fayda vardır. 

Y ekseninin işaretlerini de de, trigonometrik fonksiyonun görüntü aralığı olan [-1,1] aralığında belirlemekte fayda vardır. Sıfırı da dahil ederek üç işaret atayalım. 

Bu amaç için plt.xticks([İşaretler]) ve plt.yticks([İşaretler]) komutlarını kullanırız. Parantez içlerine, eksen işaretlerini sıraladığımız listeyi koymalıyız.



```python
plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")
plt.plot(X, S, color="green", linewidth=1.0, linestyle="-")
plt.xlim(X.min()*1.1, X.max()*1.1)  
plt.ylim(C.min()*1.1, C.max()*1.1) 
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, +1])
plt.show()
```

Görüleceği üzere grafiğimiz üzerinde, X eksenindeki pi sayısının katları sayı olarak görünmektedir. Radyal gösterimi korumak adına bu etiketlerin yazım biçimini de değiştirmek gerekmektedir. Bunun için LaTeX dökümanlama sistemininin dilini kullanarak daha teknik ve rafine bir görünüm elde edeceğiz. LaTeX sistemini öğrenmek istediğinizde, internette çok fazla sayıda kaynağa erişebilirsiniz. Matematiksel ifadelerin LaTeX gösterimleri için [bu bağlantıyı](http://www.math.ubc.ca/~cautis/tools/latexmath.html) kullanabilirsiniz. 

```python
plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")
plt.plot(X, S, color="green", linewidth=1.0, linestyle="-")
plt.xlim(X.min()*1.1, X.max()*1.1)  
plt.ylim(C.min()*1.1, C.max()*1.1) 
# plt.xticks fonksiyonuna yeni argüman ekliyoruz ve LaTeX dilinde eksen işaretlerine matematiksel görünüm kazandırıyoruz.
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
# plt.yticks fonksiyonuna yeni argüman ekliyoruz ve x ekseni ile uyumlu olması adına buradaki eksen işaretlerini de LaTeX dilinde yazıyoruz.
plt.ylim(C.min()*1.1,C.max()*1.1)
plt.yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$'])
plt.show()
```

### 2.1.5 Lejant Oluşturma (legend)
Görüldüğü üzere, örneğimizdeki iki çizimde de hangi çizginin kosinüs fonksiyonuna, hangi çizginin sinüs fonksiyonuna ait olduğunu grafiği okuyan kişiye aktaramıyoruz. Dolayısıyla, `plt.legend() `komutu kullanarak bir lejant eklememiz gerekmektedir.

plt.legend() fonksiyonunda kullanabileceğimiz çeşitli parametreler bulunmaktadır.
loc argümanı, lejantın çizim içinde nerede kullanılacağını belirlememizi sağlar. Default olarak, matplotlib en iyi seçimi yapar. Ancak bu parametreyi değiştirebiliriz. Aşağıdaki tabloda, lejantın yerini belirtebileceğimiz string ve sayı kodları bulunmaktadır.  


| Loc String     | Loc Code           | Türkçesi      |
|--------------|----------------|----------------|
| `'best'`    | 0 |En iyi seçenek (default) |
| `'upper right'`    | 1 | Sağ üst |
| `'upper left'`    | 2| Sol üst |
| `'lower left'`   | 3  |Sol alt |
| `'lower right'` | 4   | Sağ alt |
| `'right' `   | 5| Sağ |
| `'center left'`  | 6   | Sol orta |
| `'center right'`      | 7    | Sağ orta |
| `'lower center'`    | 8   | Alt orta |
| `'upper center'`     | 9   | Üst orta |
| `'center' `    | 10  | Grafiğin merkezi|

Lejantın bir çerçeve içerisinde olması arzu edilmediğinde, `frameon = False` parametresi eklenir. Lejanta kullanılacak başlık (`title`), font büyüklüğü (`fontsize`), gölgelendirme (`shadow`), lejant çerçevesinin şeffaflığı (`framealpha`), lejant çerçevesinin rengi (`edgecolor`) gibi çok daha fazla parametre de eklenebilir. 

```python
# Öncelikle plt.plot() fonksiyonuna, o grafiğin adını belirte label argümanını yerleştirelim.
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cos(x)")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-", label="sin(x)")
# Lejantı da ekleyelim. Üst solda bulusun. Lejantın çerçevesi olmasın. 
plt.legend(loc='upper left', frameon=False) 
# Grafiği ekranda gösterelim.
plt.show()
```

### 2.1.6 Eksen isimlerini oluşturma (label)
Grafiğimizin eksenlerinin bir isminin olması, okuyucu tarafından hangi eksende ne tür verinin bulunduğunu anlaması açısından önemlidir. Örneğimizdeki grafikte X ekseninde fonksiyonların x değerleri, Y ekseninde de fonksiyonların y değerleri bulunmaktadır. Bunları göstermek için `plt.xlabel()` ve `plt.ylabel()` komutları kullanılır. Bu eksen isimlerini çeşitlendirmek için de çok fazla sayıda parametre kullanılabilir. 

```python
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cos(x)")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-", label="sin(x)")
plt.legend(loc='upper left', frameon=False) 
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### 2.1.7 Grafik Başlığı Oluşturma (title)
Grafiğimizde temel bir eksiklik daha bulunmaktadır. Grafiğin neyi anlattığını belirten bir başlık bulunmamaktadır. Bu eksikliği giderebilmek için `plt.title()` komutu kullanılmaktadır. Grafik başlığı default olarak merkeze ortalanır. Bunu değiştirmek için komut içine `loc` parametresi eklenerek `'center'`, `'left'`, `'right'` stringleri kullanılabilir. Grafik başlığının, grafiğin en tepesinden uzaklığı default olarak 6 birimdir. Bunu artırmak ya da azaltmak için de `pad `parametresini kullanarak, istediğimi float değeri belirleyebiliriz. Örneğimizle pekiştirelim

```python
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cos(x)")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-", label="sin(x)")
plt.legend(loc='upper left', frameon=False) 
plt.xlabel('x')
plt.ylabel('y')
plt.title('Trigonometrik Fonksiyonlar', loc = 'left', pad = 12) # Grafiğin en tepesinden 12 birim uzaklıkta bulunan ve sola dayalı bir başlık oluşturalım.
plt.show()
```

## 2.2 Saçılım Grafiği (Scatter Plot)
Grafikler sadece çizgi grafiklerden ibaret değildir. Şimdi, verileri daire şekilli olarak grafik üstüne saçılmış scatter plot örneklerine yoğunlaşalım.
Saçılım grafiği çizdirmek için `plt.scatter(X,Y)` komutu kullanılır.

```python
import matplotlib.pyplot as plt #Kütüphaneyi çalışma ortamımıza yükleyelim.
plt.style.use('default') # Stil olarak default stili kullanalım.

yil= [1950, 1970, 1990, 2010] # Yıl listesini girelim.
nufus = [2.519, 3.692, 5.263, 6.972]  # O yıllara ait dünya nüfusu verisi listesini oluşturalım. 
plt.scatter(yil, nufus) 
plt.show()
```

Şimdi, Dünya Bankası'nın tüm ülkelere ilişkin yaşam beklentisi ve kişi başına düşen gelir verilerini (2007) kullanarak bir analiz yapmaya çalışalım. Aşağıda bu verinin tablo halinin bir kesitinin görüntüsü gösterilmiştir. ![figure](https://i.hizliresim.com/k7gJ8U.png)

Şimdilik bu dosyanın nasıl kullanacağımız listeler haline getirildiği ile ilgilenmiyoruz. 

```python
#@title Verinin hazırlanması için kodlar. 

import pandas as pd
dunya = pd.read_csv("gapminder.csv")
kbdg = pd.Series.to_list(dunya['gdp_cap'])
yb = pd.Series.to_list(dunya['life_exp'])
ulke = pd.Series.to_list(dunya['country'])
kita = pd.Series.to_list(dunya['cont'])
nufus = pd.Series.to_list(dunya['population'])

```

Elimizde aşağıdaki listeler bulunmaktadır:

`ulke` listesi: ülkelerin isimleri.

`kita `listesi: ülkelerin bulunduğu kıtaların isimleri.

`nufus` listesi: ülkelerin nüfuslarını içeren liste.

`kbdg` listesi: ülkelerin kişi başına düşen gelir verilerini içeren liste. 

`yb` listesi: ülkelerdeki kişilerin ortalama yaşam beklentisi.

Örneğin ikisini yazdıralım: 


```python
print(nufus)
```

```python
print(ulke)
```

Şimdi X ekseninde kbdg, Y ekseninde yb listelerindeki verilerin olduğu bir çizgi grafiği çizelim. 

```python
plt.plot(kbdg,yb)
plt.show()
```

Grafik berbat görünüyor. Çünkü çizgi grafikler belirli fonksiyon ya da zaman aralığında ilerleyen veriler için oldukça kullanışlıdır. Zaman ya da fonksiyon serisi olmayan, sadece birbiriyle ilişkili iki listenin olduğu verileri grafiğe aktarırken saçılım grafiğini kullanmak daha anlamlıdır. 

```python
plt.scatter(kbdg,yb)
plt.show()
```

Evet şimdi daha anlamlı bir grafik elde ettik. Ancak bu grafik, tam anlamıyla bir grafiğin tüm elemanlarına sahip değil. 
Grafiğin eksenlerinin lineer olarak düzenlendiğini gözlemleyin. Ancak veri setimizde kişi başına düşen gelir listesindeki değerlerin çok büyük bir kısmı 10000 dolar altında yoğunlaşmıştır. Bu durumda, bir yere yoğunlaşmış verilen gözlemlenmesi daha zordur. Dolayısıyla, x eksenini lineer olarak değil, logaritmik olarak düzenlemekte fayda vardır. 

```python
plt.scatter(kbdg,yb)
plt.xscale('log') 

# Aynı zamanda eksenlerin isimlerini ve grafiğin ismini de verelim.
plt.xlabel('Kişi Başına Düşen Gelir ($)')
plt.ylabel('Ortalama Yaşam Beklentisi (Yıl)')
plt.title('Gelir - Ömür İlişkisi (Dünya Bankası 2007 Kalkınma Verileri)')
plt.xticks([1000, 5000, 10000, 20000, 50000],['1k', '5k', '10k', '20k', '50k'])
plt.show()
```

Daha güzel bir grafik elde ettik. Ancak, grafiğimiz daha fazla bilgiyi barındırmalı. Saçılım grafiğindeki dairelerin boyutunu değiştirme olanağımız vardır. Bunun için `plt.scatter(X,Y,s=boyut_değeri)` komutunu kullanırız. Grafiği daha da ilginç hale getirmek için, her bir dairenin boyutunu ülke nüfusu ile orantılı bir şekilde yeniden boyutlandıralım.

```python
# NumPy kütüphanesini yükleyelim
import numpy as np

# nufus listesini bir NumPy dizisi olarak saklayalım: np_nufus
np_nufus = np.array(nufus)
# np_nufus dizisindeki milyonlarla ifade edilen nüfus verisini 1 milyon ile bölelim. Aksi takdirde daireler aşırı büyük olacağı için görünmeyecektir.
np_nufus = np_nufus / 1000000


# Grafiği oluşturalım ve s değerini nufus dizisine eşitleyelim
plt.scatter(kbdg,yb, s = np_nufus)

# Diğer parametreleri aynı bırakalım
plt.xscale('log') 
plt.xlabel('Kişi Başına Düşen Gelir ($)')
plt.ylabel('Ortalama Yaşam Beklentisi (Yıl)')
plt.title('Gelir - Ömür İlişkisi (Dünya Bankası 2007 Kalkınma Verileri)')
plt.xticks([1000, 5000, 10000, 20000, 50000],['1k', '5k', '10k', '20k', '50k'])
plt.show()
```

Bu grafikte dikkat çeken iki büyük dairenin, dünyanın en kalabalık iki ülkesine ait olduğunu biliyoruz. Dolayısıyla, bu ülkelerin isimlerini dairelerinin üzerine konumlandıralım.

## 2.3 Histogram

Histogramlar verinin nasıl dağıldığını, nerelerde yoğunlaştığını ya da seyreltiğini anlamamıza yarayan grafiklerdir. Veriler belirli aralıklara (kategorilere) ayrılır. Her bir aralık için bir bar çizilir. Barın yüksekliği, veri setinin o aralıkta daha fazla değere sahip olduğunu gösterir.
Histogram çizdirmek için plt.hist(X) komutu kullanılır. Histogram üzerinde farklı parametreler de uygulanabilmektedir.

Ortalama yaşam beklentisi dizisini kullanarak, ülkelerin yaşam beklentisi dağılımını histogram çizerek gösterelim. 

```python
import matplotlib.pyplot as plt
plt.hist(yb)
plt.show()
```

Histograma baktığımızda default olarak matplotlib'in veriyi 10 adet aralığa böldüğünü görmekteyiz. Verinin kaç aralığa bölündüğünü belirten sayıya `bin` adı verilmektedir. Bin miktarını kendimiz de belirleyebiliriz. Bunun için `plt.hist(X,bin)` komutunda, bin sayısını belirtmemiz gerekir. Örneğin aynı histogramı 5 ve 20'şer aralıkla çizdirelim.

```python
plt.hist(yb,5)
plt.show()
```

```python
plt.hist(yb,20)
plt.show()
```

Histogram çizdirirken de, daha önce bahsedilen eksen isimleri, renk, kalınlık, başlık vb. tüm parametreler de kullanılabilmektedir.
