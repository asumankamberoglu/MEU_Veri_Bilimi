## Örnek Uygulama (Kadınların Lisans Eğitimine Katılımı) 
1970-2011 yılları arasında ABD'de kadınların üniversitelerin lisans programlarına kayıt olma oranlarına ilişkin `lisans_egitimi.csv` veri setini kullanacağız. Öncelikle pandas ve NumPy kütüphanelerini kullanarak kullanacağımız verileri hazırlayalım.

```python
# Gerekli kütüphanelerin yüklenmesi
import pandas as pd
import numpy as np
# Veri setinin okunması: lisans
lisans = pd.read_csv('lisans_egitimi.csv')
# Lisans veri setinden "Yil" sütununun veri çerçevesi olarak ayıklanması: yil_df
yil_df = lisans.loc[:,"Yil"]
# Lisans veri setinden "Bilgisayar" sütununun veri çerçevesi olarak ayıklanması: bilgisayar_df
# Aynı işlemler "Fizik", "Saglik" ve "Egitim" sütunları için de tekrarlanacaktır. 
bilgisayar_df = lisans.loc[:,"Bilgisayar"]
fizik_df = lisans.loc[:,"Fizik"]
saglik_df = lisans.loc[:, "Saglik"]
egitim_df = lisans.loc[:,"Egitim"]
# Elimizdeki veri çerçevelerini np.array fonksiyonu ile NumPy dizisine çevirelim. 
yil = np.array(yil_df)
fizik = np.array(fizik_df)
bilgisayar = np.array(bilgisayar_df)
saglik = np.array(saglik_df)
egitim = np.array(egitim_df)
```

```python
# Matplotlib kütüphanesini import edelim. 
import matplotlib.pyplot as plt
# Grafik çerçevesine çizilecek olan ilk grafik fizik dizisi olsun. Bunu mavi ile gösterelim. 
plt.plot(yil, fizik, color='blue')
# Aynı çerçeveye çizilecek olan ikinci grafik bilgisayar bilimleri dizisi olsun. Bunu da kırmızı ile gösterelim. 
plt.plot(yil, bilgisayar, color='red')
# Tüm ayarları girdikten sonra grafiğimizi ekrana verelim. 
plt.show()
```

Grafikleri oluştururken tek bir çerçeve kullanmak zorunda değiliz. Grafik çerçevesini birden fazla grafikten oluşan alt çerçevelere de ayırabiliriz. Bunun için `plt.subplot()` fonksiyonu kullanılır. Bu fonksiyonun argümanları sırasıyla **(çerçeve satırı sayısı, çerçeve sütunu sayısı, grafik sırası)** şeklindedir. Grafik sırası ise her satırda sol sütundan başlar. O satırdaki sütunlar sona erdiğinde bir alttaki satıra geçerek devam eder. 

```python
# Tek satır, iki sütundan oluşan bir çerçeve oluşturalım ve birinci grafiği fizik dizesi ile çizdirelim. Rengi mavi olsun. 
plt.subplot(1, 2, 1)
plt.plot(yil, fizik, color='blue')
# Bu alt grafiğin ismini verelim. 
plt.title('Fizik')

# Şimdi de bu ikili çerçevenin ikinci grafiğini bilgisayar dizisi ile çizdirelim. Rengi kırmızı olsun. Başlığını da belirtelim. 
plt.subplot(1, 2, 2)
plt.plot(yil, bilgisayar, color='red')
plt.title('Bilgisayar Bilimleri')

# plt.tight_layout() fonksiyonu ile alt çerçeve grafiklerin arasındaki mesafeyi otomatik olarak iyileştirebilmekteyiz. 
plt.tight_layout()

# Nihai grafik çerçevesini ekrana verelim. 
plt.show()
```

```python
# Şimdi de dört alt grafikli bir uygulama yapalım. subplot fonksiyonu ile 2 satır ve 2 sütundan oluşan 4 alt grafik oluşturalım.
# Sol üstteki ilk grafik fizik dizisi ile mavi renkte oluşsun. Grafik sıra numarası = 1
plt.subplot(2, 2, 1)
plt.plot(yil, fizik, color='blue')
plt.title('Fizik')

# Sağ üstteki grafik bilgisayar dizisi ile kırmızı renkte oluşsun. Grafik sıra numarası = 2
plt.subplot(2, 2, 2)
plt.plot(yil, bilgisayar, color='red')
plt.title('Bilgisayar Bilimleri')

# Sol alttaki grafik saglik dizisi ile yeşil renkte oluşsun. Grafik sıra numarası = 3
plt.subplot(2, 2, 3)
plt.plot(yil, saglik, color='green')
plt.title('Sağlık Bilimleri')

# Sağ alttaki grafik egitim dizisi ile sarı renkte oluşsun. Grafik sıra numarası = 4
plt.subplot(2, 2, 4)
plt.plot(yil, egitim, color='yellow')
plt.title('Eğitim Bilimleri')

# Alt grafikler arası mesafeleri otomatik olarak iyileştirelim ve nihia çerçeveyi ekrana verelim. 
plt.tight_layout()
plt.show()
```

Ürettiğimiz grafikleri çalışma klasörümüze kaydetmemiz de olanaklıdır. 

```python
# bilgisayar ve fizik dizilerini tek bir grafikte farklı renklerle çizdirelim. 
plt.plot(yil,bilgisayar, color='red') 
plt.plot(yil, fizik, color='blue')

# Eksen isimlerini tayin edelim. 
plt.xlabel('Yıl')
plt.ylabel('Lisans Derecesi Kazanan Kadınlar (%)')

# x ekseninin sınırlarını (limits) belirleyelim. 
plt.xlim(1990,2010)

# y ekseninin sınırlarını (limits) belirleyelim. 
plt.ylim(0,50)

# Grafiğimize bir başlık ismi verelim. 
plt.title('Lisans Derecesi Kazanan Kadınlar (1990-2010)\nBilgisayar Bilimleri (kırmızı)\nFizik (mavi)')
plt.tight_layout()

# Grafiğimizi 'fizik_vs_bilgisayar.png' ismiyle çalışma dosyasına kaydedelim. PNG format ve 100 dot per inch (DPI) çözünürlük
plt.savefig('fizik_vs_bilgisayar.png', dpi = 100)

# Nihai grafiği ekrana verelim. 
plt.show()
```

Grafikte lejant eksikliği göze çarpmaktadır. Hemen oluşturalım: 

```python
# bilgisayar dizisini kırmızı renkle çizdirelim. Etiket olarak (label) 'Bilgisayar Bilimleri' yazalım. Lejantta böyle görünsün. 
plt.plot(yil, bilgisayar, color='red', label='Bilgisayar Bilimleri') 

# fizik dizisini mavi renkle çizdirelim. Etiket olarak (label) 'Fizik' yazalım. Lejantta böyle görünsün.
plt.plot(yil, fizik, color='blue', label='Fizik')

# Grafiğin alt orta kısmına lejantı yerleştirelim. 
plt.legend(loc='lower center')

# Eksen isimlerini (label) ve grafik başlığını(title) tayin ettikten sonra, nihai grafiği ekrana verelim.
plt.xlabel('Yıl')
plt.ylabel('Kayıt (%)')
plt.title('Kadınların Lisans Eğitim Kayıtları Dağılımı')
plt.show()
```

Grafikte önemli veya kritik bir noktayı işaret etmeyi arzu edebiliriz. Örneğin bilgisayar bilimlerine kadınların en çok oranda kayıt yaptırdığı yeri, ok işareti ile göstermeyi ele alalım. Öncelikle veri setinde bu maksimum değeri ve bu değerin hangi yıla tekabül ettiğini araştıralım. 

```python
# bilgisayar dizisindeki maksimum değeri bulalım
bilgisayar_max = bilgisayar.max()

# Bu maksimum değerin hangi yıla karşılık geldiğini NumPy kütüphanesinin .argmax() yöntemi ile filtreleyelim. 
yil_max = yil[bilgisayar.argmax()]

print(bilgisayar_max)
print(yil_max)
```

```python
# Grafiği yine baştan aynı şekilde çizdirelim. 
plt.plot(yil, bilgisayar, color='red', label='Bilgisayar Bilimleri') 
plt.plot(yil, fizik, color='blue', label='Fizik')
plt.legend(loc='lower center')

# 1983 yılındaki %37.1'lik oranı gösteren siyah oku oluşturalım. Bunun için plt.annotate fonksiyonunu kullanacağız. 
# İlk argüman okun üzerindeki metni içeren string olacaktır: 'Maksimum Oran'
# İkinci argüman siyah okun ucunun grafikte hangi koordinata konulacağını belirlediğimiz xy parametresini girdiğimiz yerdir.
# Üçüncü argüman ise okun üzerindeki metnin başladığı koordinatı belirlediğimiz parametredir.
# Son argüman ise ok parametrelerini girdiğimiz sözlüktü. Burada sadece okun yüzeyinin rengini facecolor anahtar sözcüğü ile
# siyah (black) olarak girdik. Başka özellikler de peşi sıra eklenebilir. 
plt.annotate('Maksimum Oran', xy = (yil_max,bilgisayar_max), xytext = (yil_max+4, bilgisayar_max+4), arrowprops=dict(facecolor='green'))

# Eksen isimleri ve grafik başlığını girip, grafiği ekrana verelim. 
plt.xlabel('Yıl')
plt.ylabel('Kayıt (%)')
plt.title('Kadınların Lisans Eğitim Kayıtları Dağılımı')
plt.show()
```
